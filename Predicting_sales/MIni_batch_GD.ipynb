```python
df = pd.read_csv("advertising.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
new_df = scaler.fit_transform(df.drop(columns=('Sales')))
new_df = pd.DataFrame(new_df,columns=('a','b','c'))
new_df
```

```python
y = df['Sales'].values
X = new_df.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X
```

```python
import random
class MGD:
    def __init__(self,epochs,lr,batch_size):
        self.coef = None
        self.intercept = None
        self.epochs = epochs
        self.lr = lr
        self.batch_size = batch_size
    def fit(self,X_train,y_train):
      self.coef = np.ones(X_train.shape[1])
      self.intercept = 0
        # print(self.coef.shape)
      for i in range(self.epochs):
       for j in range(int(X_train.shape[0]/self.batch_size)):   
         batch = random.sample(range(X_train.shape[0]),self.batch_size)
         y_hat = np.dot(X_train[batch],self.coef)+self.intercept
         # print(y_hat)   
         slope_i = -2*(np.mean(y_train[batch]-y_hat)) 
         slope_c = -2*(np.dot((y_train[batch]-y_hat),X_train[batch]))/self.batch_size   
         # print(slope_c.shape)
         self.coef = self.coef - (self.lr*slope_c)    #without scaling this line raises an error
         self.intercept = self.intercept - (self.lr*slope_i)       
      print(self.intercept,self.coef)
    def predict(self,X_test):
        return np.dot(X_test,self.coef) + self.intercept
```

```python
mgd = MGD(50,0.01,10)
mgd.fit(X_train,y_train)
pred = mgd.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```
