```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
new_df = scaler.fit_transform(df.drop(columns=('Sales')))
new_df = pd.DataFrame(new_df,columns=['a','b','c'])
new_df
```

```python
y = df['Sales']
X = new_df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
class GD:
    def __init__(self,epochs,lr):
        self.coef = None
        self.intercept = None
        self.epochs = epochs
        self.lr = lr
    def predict(self,X_train,y_train):
        self.coef = np.ones(X_train.shape[1])
        self.intercept = 0
        # print(self.coef.shape)
        for i in range(self.epochs):  
         y_hat = np.dot(X_train,self.coef)+self.intercept
         # print(y_hat)   
         slope_i = -2*(np.mean(y_train-y_hat)) 
         slope_c = -2*(np.dot((y_train-y_hat),X_train))/X_train.shape[0]   
         # print(slope_c.shape)
         self.coef = self.coef - (self.lr*slope_c)
         self.intercept = self.intercept - (self.lr*slope_i)       
        print(self.intercept,self.coef)
```

```python
gd = GD(100,0.1)
gd.predict(X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
PRED = LR.predict(X_test)
print(LR.coef_)
print(LR.intercept_)
```
